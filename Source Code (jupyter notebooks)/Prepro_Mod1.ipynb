#### Merging surface data and pressure data

````python

Mod1 = pd.merge(sdf, pdf, on=['valid_time', 'latitude', 'longitude'], how= 'inner')
````

````python
Mod1.head(7)
````

````python
Mod1.info()
````

````python
# Wind Speed Magnitude (Pythagorean theorem)

Mod1['wind_speed'] = np.sqrt(Mod1['u10']**2 + Mod1['v10']**2) 
````

````python
# Temperature Gradient (Upper temp minus Lower temp)

Mod1['temp_gradient'] = Mod1['950 hpa'] - Mod1['1000 hpa']
````

````python
# Extracting 'hour' (0-23) and 'month' (1-12)

Mod1['hour'] = Mod1['valid_time'].dt.hour
Mod1['month'] = Mod1['valid_time'].dt.month
````

````python
# Lagged blh and wind data
# Shift down by 1 row (representing a 6-hour lag)

Mod1['blh_lag_6h'] = Mod1['blh'].shift(1)
Mod1['wind_lag_6h'] = Mod1['wind_speed'].shift(1)
````

````python
#cleaning

Mod1.drop(columns = ['u10', 'v10', '950 hpa', '1000 hpa', 'latitude', 'longitude'], inplace=True)

#removed a null row
````

````python
Mod1 = Mod1.set_index('valid_time')
````
